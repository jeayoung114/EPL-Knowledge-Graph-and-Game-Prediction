```python
import pandas as pd
import fileinput
import json
```

```python
world_game = pd.DataFrame()
with fileinput.input(files="game_worldfootball.jl") as file:
    for line in file:
        conv = json.loads(line)
        world_game = world_game.append(conv, ignore_index=True)
world_game.head()
```

```python
def f(x):
    return str(x.season) + str(int(x["round"]) + 100)
world_game["time"] = world_game.apply(lambda x : f(x), axis = 1)
```

```python
time_list = sorted(list(world_game["time"].drop_duplicates()))
```

```python
time2idx = dict()
for idx, time in enumerate(time_list):
    time2idx[time] = idx
```

```python
world_game["timestamp"] = world_game['time'].apply(lambda x : time2idx[x])
```

```python
world_game
```

```python
world_game = world_game.sort_values(by = "timestamp")[["timestamp", "home_team", "away_team", "score", "season", "round", "date"]]
```

```python
world_game.to_csv("TKG_input_data.csv")
```

```python
group_dict = []
team2idx = dict()
idx2team = dict()
for idx, team in enumerate(world_game["away_team"].drop_duplicates()):
    team_dict = dict()
    team_dict['id'] = idx
    team_dict["content"]  = team
    group_dict.append(team_dict)
    team2idx[team] = idx
    idx2team[idx] = team
```

```python
with open("groups.json", "w") as f:
    json.dump(group_dict, f)
```

```python
def f(x):
    tmp = x.split(":")
    try:
        if int(tmp[0]) > int(tmp[1]):
            return "win"
        elif int(tmp[0]) < int(tmp[1]):
            return "lose"
        else:
            return "draw"
    except:
        return "-"
world_game["result"] = world_game["score"].apply(lambda x: f(x))
```

```python
future = pd.read_csv("future.csv")
```

```python
future = future.rename({"score" : "result"}, axis =1 )
```

```python
future
```

```python
world_game_future = world_game.merge(future, left_on = ["timestamp", "home_team", "away_team"], right_on = ["timestamp", "home_team", "away_team"], how = "outer")
world_game_future
```

```python
def f(x):
    if type(x.result_y) == float:
        return x.result_x
    else:
        return x.result_y

world_game_future['res'] = world_game_future.apply(lambda x : f(x), axis = 1)
```

```python
world_game_future.tail(20)
```

```python
result_json = []
for idx, row in world_game_future.iterrows():
    game_json = dict()
    game_json["id"] = idx
    game_json["group"] = team2idx[row.home_team]
    game_json['content'] = "vs " + row.away_team + " : " + row.result
    
    game_json2 = dict()
    game_json2["id"] = idx + 100000
    game_json2["group"] = team2idx[row.away_team]
    if row.result == "win":
        tmp = "lose"
    elif row.result == "lose":
        tmp = "win"
    else:
        tmp = "draw"
    game_json2['content'] = "vs " + row.home_team + " : " + tmp
    
    
    result_json.append(game_json)
    result_json.append(game_json2)
    
    
```
